```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
school_data_df = pd.DataFrame(student_data)
#print(school_data_df.head())
#print(student_data.head())

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

```python
total_schools = len(school_data_complete["school_name"].unique())
total_students = len(school_data_complete["student_name"])
#school_data_complete["budget"]
math_scores_avg = round(school_data_complete["math_score"].mean(),6)
reading_scores_avg = round(school_data_complete["reading_score"].mean(),5)
overall_passing_rate = round((math_scores_avg + reading_scores_avg) / 2,6)
passing_math = school_data_complete["math_score"] >= 70
per_passing_math = round(passing_math.sum() / total_students *100,6)
passing_reading = school_data_complete["reading_score"] >= 70
per_passing_reading = round(passing_reading.sum() / total_students * 100,6)
total_budget = school_data_complete["budget"].unique().sum()
summary_table = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students],
                              "Total Budget": [total_budget], "Average Math Score": [math_scores_avg],
                              "Average Reading Score": [reading_scores_avg], "% Passing Math": [per_passing_math],
                               "% Passing Reading": [per_passing_reading], 
                               "% Overall Passing Rate": [overall_passing_rate]
                              })

#print(math_scores_avg)
#print(reading_scores_avg)
summary_table
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
performance = school_data_complete[["school_name", "type", "student_name", "budget", "math_score", "reading_score"]]
#to group the school and get total student count then make a schools and students table
school_group = performance.groupby(["school_name"]) 
school_group_df = pd.DataFrame(school_group["student_name"].count())
school_group_df.reset_index(inplace = True)
schools_n_students = school_group_df[["school_name", "student_name"]]
schools_n_students_df = schools_n_students.rename(columns={"school_name": "School Name", "student_name": "Student Count"})

#avg math and reading score per school
avg_math = pd.DataFrame(school_group["math_score"].mean())
avg_math.reset_index(inplace = True)
avg_math_by = avg_math[["school_name", "math_score"]]
avg_math_by_school = avg_math_by.rename(columns={"school_name": "School Name", "math_score": "Average Math Score"})

avg_reading = pd.DataFrame(school_group["reading_score"].mean())
avg_reading.reset_index(inplace = True)
avg_reading_by = avg_reading[["school_name", "reading_score"]]
avg_reading_by_school = avg_reading_by.rename(columns={"school_name": "School Name", "reading_score": "Average Reading Score"})

#to group the school and get only passing math students then make a table 
per_math_grades = performance.loc[performance["math_score"] >= 70]
passing_math_group = per_math_grades.groupby(["school_name"]) 
passed_math = pd.DataFrame(passing_math_group["student_name"].count())
passed_math.reset_index(inplace = True)
passed_math_by_school = passed_math[["school_name", "student_name"]]
passed_math_by_school_df = passed_math_by_school.rename(columns={"school_name": "School Name", 
                                                                 "student_name": "Passed Math Class Count"})

#to group the school and get only passing reading students then make a table
per_reading_grades = performance.loc[performance["reading_score"] >= 70]
passing_reading_group = per_reading_grades.groupby(["school_name"]) 
passed_reading = pd.DataFrame(passing_reading_group["student_name"].count())
passed_reading.reset_index(inplace = True)
passed_reading_by_school = passed_reading[["school_name", "student_name"]]
passed_reading_by_school_df = passed_reading_by_school.rename(columns={"school_name": "School Name",
                                                                     "student_name": "Passed Reading Class Count"})
#schools_n_students = school_group["school_name", "student_name"]
#schools_n_students.reset_index(inplace=True)
#schools_unique = school_group.count().sort_values(["student_name"], ascending = False)

#per_list = pd.merge(schools_n_students, )
#school_group.count().head()
#passing_group.count()
#per.head()
#performance.head()
#schools_n_students_df
#avg_reading_by_school
#passing_math_group.mean()
#passed_reading_by_school_df
```

```python
type_budget = school_data_complete[["school_name", "type", "budget"]].drop_duplicates().rename(columns={"school_name": "School Name",
                                                                                               "type": "Type",
                                                                                               "budget": "Budget"})
#type_budget.sort_values("Budget", ascending = False)
```

```python
mg_avg_math = pd.merge(schools_n_students_df, avg_math_by_school, how="left", 
                                      on = ["School Name", "School Name"])
mg_avg_read = pd.merge(mg_avg_math, avg_reading_by_school, how= "left",
                      on = ["School Name", "School Name"])
mg_passed_math = pd.merge(mg_avg_read, passed_math_by_school_df, how= "left",
                         on = ["School Name", "School Name"])
mg_passed_reading = pd.merge(mg_passed_math, passed_reading_by_school_df, how="left", 
                                      on = ["School Name", "School Name"])
mg_total = pd.merge(type_budget, mg_passed_reading, how= "inner",
                   on= ["School Name", "School Name"])
#mg_total

#add Per Student Budget
mg_total["Per Student Budget"] = mg_total["Budget"] / mg_total["Student Count"]

#add percent passing math
mg_total["% Passing Math"] = round(mg_total["Passed Math Class Count"] / mg_total["Student Count"] *100, 6)

#add percent passing reading
mg_total["% Passing Reading"] = round(mg_total["Passed Reading Class Count"] / mg_total["Student Count"] *100, 6)

#add percent overall passing
mg_total["% Overall Passing Rate"] = round((mg_total["% Passing Math"] 
                                           + mg_total["% Passing Reading"]) / 2, 6)
#mg_total.sort_values("% Overall Passing Rate", ascending = False).reset_index().head()
```

```python
top_by = mg_total[["School Name", "Type", "Student Count", "Budget", "Per Student Budget", "Average Math Score",
                  "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]

top_schools = top_by.sort_values("% Overall Passing Rate", ascending = False)
top_by_performance_schools = top_schools.rename(columns={"Type": "School Type", "Student Count": "Total Students",
                                          "Budget": "Total School Budget"})
top_by_performance_schools.set_index("School Name").head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
nine_m = school_data_complete.loc[(school_data_complete["grade"] == "9th"), 
                                 ["school_name", "student_name","math_score"]].rename(columns={"math_score": "9th"})
ten_m = school_data_complete.loc[(school_data_complete["grade"] == "10th"), 
                                 ["school_name", "student_name","math_score"]].rename(columns={"math_score": "10th"})
eleven_m = school_data_complete.loc[(school_data_complete["grade"] == "11th"), 
                                 ["school_name", "student_name","math_score"]].rename(columns={"math_score": "11th"})
twelve_m  = school_data_complete.loc[(school_data_complete["grade"] == "12th"), 
                                 ["school_name", "student_name","math_score"]].rename(columns={"math_score": "12th"})

ninth_score_m = nine_m.groupby("school_name").mean().reset_index()
tenth_score_m = ten_m.groupby("school_name").mean().reset_index()
eleventh_score_m = eleven_m.groupby("school_name").mean().reset_index()
twelfth_score_m = twelve_m.groupby("school_name").mean().reset_index()


comb_1_m = pd.merge(ninth_score_m, tenth_score_m, how= "left", on = "school_name")
comb_2_m = pd.merge(comb_1_m, eleventh_score_m, how= "left", on = "school_name")
math_by_grade = pd.merge(comb_2_m, twelfth_score_m, how= "left", on = "school_name")
math_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
spending_groups = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
student_count = school_data_complete[["school_name", "student_name"]]
student_count = student_count.groupby("school_name").count().reset_index()


school_budget = school_data_complete[["school_name", "budget"]].drop_duplicates()
#school_budget.head()
student_count.head()

per_student_spending = pd.merge(student_count, school_budget, 
                                how = "left", on = "school_name").rename(columns={"student_name": "student count"})
per_student_spending["Per Student Budget"] = per_student_spending["budget"] / per_student_spending["student count"]
#per_student_spending.head()
```

```python
spending_pass_math = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
spending_pass_math = spending_pass_math[["school_name", "math_score"]]
spending_pass_math = spending_pass_math.groupby("school_name").count().reset_index()

spending_pass_read = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
spending_pass_read = spending_pass_read[["school_name", "reading_score"]]
spending_pass_read = spending_pass_read.groupby("school_name").count().reset_index()

stu_count = school_data_complete[["school_name", "student_name"]]
stu_count = stu_count.groupby("school_name").count().reset_index()
stu_count.head()

spending_pass = pd.merge(spending_pass_math, spending_pass_read, how = "left", on="school_name")
spending_pass = pd.merge(stu_count, spending_pass, how = "left", on = "school_name")
#spending_pass
```

```python
spending_all = school_data_complete[["school_name", "math_score", 
                                     "reading_score"]].rename(columns={"math_score": "Average Math Score",
                                                                       "reading_score": "Average Reading Score"})
spending_all = spending_all.groupby("school_name").mean().reset_index()
spending_scores = pd.merge(spending_all, spending_pass, how = "left", on="school_name")
spending_scores["% Passing Math"] = spending_scores["math_score"] / spending_scores["student_name"] *100
spending_scores["% Passing Reading"] = spending_scores["reading_score"] / spending_scores["student_name"] *100
spending_scores["% Overall Passing Rate"] = (spending_scores["% Passing Math"] + spending_scores["% Passing Reading"])/2


per_student_spending["Spending Range (Per Student)"] = pd.cut(per_student_spending["Per Student Budget"],
                                                                                 spending_bins, 
                                                                                 labels = spending_groups)
per_student_spending


merge_spending = pd.merge(per_student_spending, spending_scores, how = "inner", on="school_name")
spending_summary = merge_spending.groupby("Spending Range (Per Student)").mean().reset_index().drop(
    ["student count", "budget", "Per Student Budget"],axis = 1)
spending_summary.drop(["student_name", "math_score", "reading_score"], axis = 1)
```
